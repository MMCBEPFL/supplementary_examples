```python
#import important libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets
from IPython.display import set_matplotlib_formats
#set_matplotlib_formats('png', 'pdf')
```

## Bifurcation diagram

```python
g_domain = np.linspace(0,1.5,100)

#create a widget to play with the parameters
def r(g):
    return 2*g/(g**2+1)**2

def s(g):
    g2=g**2
    return g2*(1-g2)/(g2+1)**2

plt.figure(figsize=(10,10))
plt.plot(g_domain, s(g_domain), c='blue')
plt.plot(g_domain, r(g_domain), c='red')
plt.xlabel('g')
plt.ylabel('r(g) [blue], s(g)[red]')
plt.ylim([-0.1,0.8])
plt.axhline(y=0, ls='--', c='grey')
plt.show()
```

```python
g_domain = np.linspace(0,2,100)

#create a widget to play with the parameters
def r(g):
    return 2*g/(g**2+1)**2

def s(g):
    g2=g**2
    return g2*(1-g2)/(g2+1)**2

plt.figure(figsize=(10,10))
plt.plot(r(g_domain), s(g_domain), c='blue')
plt.xlabel('r')
plt.ylabel('s')
plt.xlim([0,1])
plt.ylim([0,0.15])
plt.axhline(y=1/8, ls='--', c='grey')
plt.show()
```

## $F(g)$ _vs_ $g$

```python
g_domain = np.linspace(0,2,100)

#create a widget to play with the parameters
def F(g,s,r):
    return s-r*g+g**2/(g**2+1)
    
def f(s = 0.05, r = 0.5):
    plt.figure(figsize=(10,10))
    plt.plot(g_domain, F(g_domain,s,r))
    plt.xlabel('g')
    plt.ylabel('F(g)')
    plt.title('Interactive plot: s=' + str(s) + ' r=' + str(r))
    plt.ylim([-0.5,0.5])
    plt.axhline(y=0, ls='--', c='grey')
    plt.show()
    
interact(f, s = (0,1,0.01), r = (0,2,0.05))
```

## $f(g)$, $h(g)$ _vs_ $g$

```python
g_domain = np.linspace(0,3,100)

#create a widget to play with the parameters
def f(g,s,r):
    return -s+r*g

def h(g):
    return g**2/(g**2+1)

def ff(s = 0.05, r = 0.5):
    plt.figure(figsize=(20,10))
    plt.plot(g_domain, h(g_domain), c='black')
    plt.plot(g_domain, f(g_domain, s, r))

    plt.xlabel('g')
    plt.ylabel('f(g), h(g)[blue]')
#    plt.title('Interactive plot: s=' + str(s) + ' r=' + str(r))
    plt.ylim([-0.2,1.1])
    plt.axhline(y=0, ls='--', c='grey')
    plt.axhline(y=1, ls='--', c='grey')
    plt.show()
    
interact(ff, s = (0,1,0.01), r = (0,2,0.05))
```
