# Kuramoto model

```python
#import important libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from scipy.integrate import odeint, quad
from IPython.display import set_matplotlib_formats
from matplotlib.markers import MarkerStyle
set_matplotlib_formats('png', 'pdf')
```

```python
si_w = 1;
Kc = 2 / np.pi * np.sqrt(2*np.pi) * si_w 
def integrand(x,K,r,sig):
    return np.cos(x)**2* np.exp(-(np.sin(x)*K*r)**2/(2*sig**2))/ (np.sqrt(2*np.pi)*sig)

def integral(K,r,sig):
    return quad(integrand, -np.pi/2, np.pi/2, args=(K,r,sig))

def rel(K,r):
    return K*integral(K,r,si_w)[0]-1
```

```python
contr=np.arange(0,1.1,0.01)
contK=np.arange(0,5,0.1)
KK,rr=np.meshgrid(contK,contr)
V=np.array([0.])
Z=np.zeros_like(KK)
for i in range(KK.shape[0]):
    for j in range(KK.shape[1]):
        Z[i,j]=rel(KK[i,j],rr[i,j])
plt.contour(KK,rr,Z,V, colors='purple')
plt.plot([0,5],[0,0], color='purple')
plt.scatter(Kc,0)
plt.xlabel('K')
plt.ylabel('r')
plt.show()
```
